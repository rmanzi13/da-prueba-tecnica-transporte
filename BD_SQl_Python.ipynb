```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
file_path = '/content/Datos_Filtrados.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
data
```

```python
data.info()
```

```python
# Convertir Numero_Pasajeros a int
data['Numero_Pasajeros'] = data['Numero_Pasajeros'].astype(int)
```

```python
# Convertir fecha a datetime
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Convertir 'duracion_viaje_minutos' a float
data['Duracion_Viaje_Minutos'] = pd.to_numeric(data['Duracion_Viaje_Minutos'], errors='coerce')
```

```python
data.info()
```

```python
# Filtrando los tipos de transporte inviables ('Tranvía' y 'Metro')
transportes_inviables = ['Tranvía', 'Metro']
data_filtrada = data[~data['Tipo_Transporte'].isin(transportes_inviables)]

# Mostrando un resumen de los datos filtrados para verificar
resumen_transporte = data_filtrada['Tipo_Transporte'].value_counts()
print(resumen_transporte)
```

```python
# Filtrar retrasos anormalmente altos (999 minutos)
data_filtrada = data_filtrada[data_filtrada['Retraso_Minutos'] < 999]

print("Datos después del filtrado:")
# Mostrar una muestra de los datos filtrados
print(data_filtrada.head())
```

```python
data_filtrada.head()
```

```python
data_filtrada.info()
```

```python
# Guardar el DataFrame en un archivo CSV
file_path = '/content/data_final.csv'
data_filtrada.to_csv(file_path, index=False)
```

```python
import sqlite3

# Conectar a la base de datos SQLite (o crear una nueva)
conn = sqlite3.connect('transporte_publico.db')
c = conn.cursor()

# Crear la tabla
c.execute('''
CREATE TABLE IF NOT EXISTS transporte_publico (
    ID_Viaje INTEGER PRIMARY KEY,
    Fecha DATETIME,
    Ruta TEXT,
    Numero_Pasajeros INTEGER,
    Duracion_Viaje_Minutos FLOAT,
    Retraso_Minutos INTEGER,
    Tipo_Transporte TEXT,
    Region TEXT,
    Dia_Semana TEXT
)
''')

conn.commit()
```

```python
# Conectar a la base de datos SQLite
conn = sqlite3.connect('transporte_publico.db')
c = conn.cursor()

# Ejecutar una consulta SQL
c.execute('SELECT * FROM transporte_publico LIMIT 5')

# Obtener los resultados
rows = c.fetchall()
```

```python
# Convertir los resultados en un DataFrame de pandas
columns = [desc[0] for desc in c.description]
df = pd.DataFrame(rows, columns=columns)

# Mostrar el DataFrame
print(df)
```

```python
# Cambios en archivo para su uso en Power BI
import pandas as pd
# Cargamos los datos
file_path = '/content/data_final.csv'
data = pd.read_csv(file_path)

# Vemos las priemras filas para entender su estructura
data.head()
```

```python
# Convertiendo fecha a datetime
data['Fecha'] = pd.to_datetime(data['Fecha'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Asegurarnos que Duracion_Viaje_Minutos es float
data['Duracion_Viaje_Minutos'] = data['Duracion_Viaje_Minutos'].astype(float)

# Reemplazando punto decimal por coma para no tener conflicto en Power BI
data['Duracion_Viaje_Minutos'] = data['Duracion_Viaje_Minutos'].apply(lambda x: str(x).replace('.', ','))
```
